```python
import pandas as pd

df = pd.read_csv("G:\Datasets-master\Titanic-Dataset.csv")
```

```python
# 1 . Find out how many male and female passenger was onboarded

male_pass=[]
female_pass=[]
male_pass_cnt,female_pass_cnt=0,0

for i in df["Sex"]:
    if i=="male":
        male_pass.append(i)
        male_pass_cnt+=1
    else:
        female_pass.append(i)
        female_pass_cnt+=1
        
print(f"number of male and female passengers are = {male_pass_cnt} and {female_pass_cnt} respectively.")
```

```python
'''
2. how many survived we have .
3. how many casuality we have.
'''

survived_pass=[]
casualties=[]
survived_pass_cnt,casualties_cnt=0,0

for i in df["Survived"]:
    if i==1:
        survived_pass.append(i)
        survived_pass_cnt+=1
    else:
        casualties.append(i)
        casualties_cnt+=1
        
print(f"number of survived passengers and casualties are = {survived_pass_cnt} and {casualties_cnt} respectively.")
```

```python
# 4 . what is name of a person who is the eldest one . 


df[df["Age"]==max(df["Age"])]
```

```python
# 5 . how many passenger do we have in first , second and third class 

cnt_first_class_pass=0
cnt_second_class_pass=0
cnt_third_class_pass=0

for i in df['Pclass']:
    if i==1:
        cnt_first_class_pass+=1
    elif i==2:
        cnt_second_class_pass+=1
    else:
        cnt_third_class_pass+=1
        
        
print(f"number of passengers travelling in first,second and third class are {cnt_first_class_pass}, {cnt_second_class_pass} and {cnt_third_class_pass} respectively.")
```

```python
# 6. how many person we have whose name starts with "s"

#
count_names_starting_with_s = (df['Name'].str.startswith('S', na=False)).sum()

#
count_names_starting_with_s = df[df['Name'].str.startswith('S', na=False)].shape[0]

print(f"The number of persons whose names start with 'S' is: {count_names_starting_with_s}")


print(f"The number of persons whose names start with 'S' is: {count_names_starting_with_s}")
```

```python
# 7. try to create a new column which is a summation  of "SibSp" and "parch"

df['SibSp+Parch'] = df["SibSp"].astype(str)+" " +df["Parch"].astype(str)

df['SibSp+Parch']
```

```python
'''
8 . how many person do we have below age of 25 .
9 . how many person died whose age was less then 40
'''

cnt_of_person_below_age_25 = 0
cnt_of_person_died_above_age_40 = 0

# Assuming "Age" and "Survived" are columns in the DataFrame df
for i, j in df.iterrows():
    age_value = j['Age']
    survived_value = j['Survived']
    
    if age_value < 25:
        cnt_of_person_below_age_25 += 1
    
    if age_value > 40 and survived_value == 0:
        cnt_of_person_died_above_age_40 += 1

print(f"People below age 25: {cnt_of_person_below_age_25}")
print(f"People who died above age 40: {cnt_of_person_died_above_age_40}")
```

```python
'''
8 . how many person do we have below age of 25 .
9 . how many person died whose age was less then 40
'''

cnt_of_person_below_age_25 = 0
cnt_of_person_died_above_age_40 = 0

for i,age in enumerate(df["Age"],start=0):
    if age < 25:
        cnt_of_person_below_age_25 += 1
    elif age > 40:
        if df["Survived"][i] == 0:
            cnt_of_person_died_above_age_40 += 1

print(f"People below age 25: {cnt_of_person_below_age_25}")
print(f"People who died above age 40: {cnt_of_person_died_above_age_40}")
```
